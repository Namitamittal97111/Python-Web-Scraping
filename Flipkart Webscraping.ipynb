```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python

product_name=[]
prices=[]
description=[]
Reviews=[]


for i in range(1,10):
    url="https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    r=requests.get(url)
    print(r)


    soup=BeautifulSoup(r.text, 'lxml')
    box=soup.find('div',class_='_1YokD2 _3Mn1Gg')
    #print(soup.prettify)

    names=box.find_all('div',class_ = '_4rR01T')
    for i in names:
        name=i.text
        product_name.append(name)


    price =box.find_all('div', class_='_30jeq3 _1_WHN1')
    for i in price:
        name=i.text
        prices.append(name)


    desc=box.find_all('ul', class_='_1xgFaf')
    for i in desc:
        name=i.text
        description.append(name)


    rev=box.find_all('div', class_='_3LWZlK')
    for i in rev:
        name=i.text
        Reviews.append(name)
print(len(prices)) 
print(len(Reviews))
```

```python
df=pd.DataFrame({"Product Name":product_name,"Prices":prices, "Description":description})
df
```
